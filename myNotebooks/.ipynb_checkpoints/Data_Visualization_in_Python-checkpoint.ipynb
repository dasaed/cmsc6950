# Visualizing Data in Python 
### Using matplotlib



```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

fig, ax_lst = plt.subplots(2,3) # a figure with a 2x3 grid of axes
fig.suptitle("A figure with a 2x3 Grid of Axes")
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
fig2, (ax1,ax2) = plt.subplots(2,2)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

fig = plt.figure() # create a new figure
axis1 = fig.add_subplot(1,2,1) # just left 
axis2 = fig.add_subplot(2,2,2) # top right
axis3 = fig.add_subplot(2,2,4) # bottom right

```

```python
%matplotlib inline
from matplotlib import pyplot as plt

axis = plt.axis([2,6,4,8])
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
fig, ax_list = plt.subplots(2,2)

ax_list[0,0].set_xlim(0, 4)
ax_list[0,1].set_ylim(2, 6)
ax_list[1,0].set_xlabel("Label for x axis 10")
ax_list[1,1].set_ylabel("Label for y axis 11")

plt.tight_layout() # adjusting the location of axes
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

fig, ax_lst = plt.subplots(2,2)  # a figure with a 2x2 grid of Axes

ax_lst[0,0].plot([1, 2, 3, 4], "r")
ax_lst[0,0].bar([1,2,3],[3,4,5])
ax_lst[1,1].arrow(1,0,0.5,0.2)
ax_lst[0,1].plot([1, 2, 3, 4], [1, 8, 27, 64])
ax_lst[1,0].barh([1,2,5],[0.5,1,2])
ax_lst[1,1].axhline(0.45)

plt.tight_layout() # adjusting the location of axes
```
